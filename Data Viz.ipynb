```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Histogram of sex ratio:
figure = plt.figure(figsize=(15,8))
plt.hist(df['sex_ratio'],color = 'olive',bins = 30)
plt.xlabel('sex_ratio')
plt.ylabel('count of cities')
plt.title('Histogram of Sex-ratio')

#Results:
#The Data is almost normally distributed. Most of the cities have Sex-Ratio between 870 and 1030
#5 cities have sex ratio less than 800
#7 cities have sex ratio more than 1050
```

```python
# the distribution of Child sex ratio:
#The kernel density estimate may be less familiar, but it can be a useful tool for plotting the shape of a distribution.
#Like the histogram, the KDE plots encode the density of observations on one axis with height along the other axis:
sns.distplot(df['child_sex_ratio'],rug = True)
```

# Based on sex ratio

```python
#Top 10 cities based on sex ratio
top_10cities=df.sort_values('sex_ratio',ascending = False)[:10]
top_10cities
```

```python
#Bottom 10 cities based on sex ratio
bottom_10cities=df.sort_values('sex_ratio')[:10]
bottom_10cities
```

```python
#Appending both top and bottom 10
frames=[top_10cities,bottom_10cities]
newframe = pd.concat(frames,ignore_index = True) 
newframe
```

```python
#Lollipop chart for Top 10 and bottom 10 cities based on sex ratio
fig, ax = plt.subplots()
Colours = ['gold', 'yellowgreen']
my_color=np.where(newframe['South/North']=='South', 'orange', 'skyblue')


ax.hlines(newframe.name_of_city, xmin=0, xmax=newframe.sex_ratio,color=my_color)
ax.plot(newframe.sex_ratio, newframe.name_of_city, "o")
ax.set_xlim(0, 1300)
plt.ylabel("Top and Bottom 10 Cities")
plt.xlabel("Sex Ratio")

plt.title("Sex Ratio")
plt.legend(['North','South'] ,loc='upper right') 
plt.show()

#Loolipop with blue bar indicates NOrth indian cities and orange color indicates south indian cities
#The length of the bar indicates the sex ratio KPI
```

By studying the histogram and the above 2 tables of sex ratio the following is evident:
    1.The all states in the bottom 10 sex ratio is from the 'North India'
    2.Out of top 10 states in the sex ratio, 8 states are from the south

Why sex ratio is important?
    It gives us the insight how each sex is treated whether equally or not
    When the ratio is balanced the crime rate is less otherwise the crimes against women would be more
    In the equal sex community the women will be treated and guven equal opportunity to men

From this KPI we can clearly prove that South india is better than the north india

#               Based on Female graduate population

```python
#female graduates:
data_female_graduates = pd.DataFrame(df['female_graduates'].groupby(df['South/North']).mean().sort_values(ascending = False)[:5])
data_female_graduates
```

```python
colors_list = ['gold', 'yellowgreen']
explode_list = [0.1, 0]
data_female_graduates['female_graduates'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,  # add custom colors
                            explode=explode_list # 'explode' lowest 3 continents
                            )
 
# scale the title up by 12% to match pctdistance
plt.title('Female Graduates Population', y=1) 
 
plt.axis('equal') 
 
# add legend
plt.legend(labels=data_female_graduates.index, loc='upper left') 
 
plt.show()
```

```python
#Top 5 cities based on female graduates population
data_female_graduates_best_city = pd.DataFrame(df['female_graduates'].groupby(df['name_of_city']).mean().sort_values(ascending = False)[:5])
data_female_graduates_best_city['city_name'] = data_female_graduates_best_city.index.values
data_female_graduates_best_city 
```

```python
#Bottom 5 cities
data_female_graduates_worst_city = pd.DataFrame(df['female_graduates'].groupby(df['name_of_city']).mean().sort_values()[:5])
data_female_graduates_worst_city['city_name'] = data_female_graduates_worst_city.index.values
data_female_graduates_worst_city
```

```python
#concatebating both top and bottom cities
frames=[data_female_graduates_best_city ,data_female_graduates_worst_city]
df2 = pd.concat(frames) 
df2
```

```python
merged_df = pd.merge(df2[['female_graduates','city_name']],
                     df[['name_of_city','South/North']],
                     left_on = 'city_name', 
                     right_on = 'name_of_city', 
                     how='left')

print (merged_df)
```
