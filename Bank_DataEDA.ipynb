```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('./bank/bank.csv',sep=';')
```

```python
df.head(5)
```

```python
#Function to replace values in marital column
def replace_marital(val):
    if val=='single':
        return 0
    else:
        return 1
```

```python
#using 'apply' function to replace the values
df['marital']=df['marital'].apply(replace_marital,1)
df.head(5)
```

```python
#using 'map' function to substitute the values
df['housing']=df['housing'].map({
    'no':0,
    'yes':1
}.get)
```

```python
df.head()
```

```python
#using 'replace' method to substitue values
df['loan']=df['loan'].replace({
    'no':0,
    'yes':1
})
```

```python
df.head()
```

```python
df['job'].unique()
```

```python
df['job'].replace({
    'unknown':np.nan,
    'management':0,
    'technician':1,
    'entrepreneur':2,
    'blue-collar':3,
    'retired':4,
    'admin.':5,
    'services':6,
    'self-employed':7,
    'unemployed':8,
    'housemaid':9,
    'student':10
},inplace=True)
```

```python
df['job'].unique()
```

```python
df.head()
```

```python
df['education'].unique()
```

```python
df['education'].replace({
    'unknown':np.nan,
    'tertiary':0,
    'secondary':1,
    'primary':2
},inplace=True)
```

```python
df.head()
```

```python
df['default'].replace({
    'no':0,
    'yes':1
},inplace=True)
```

```python
df.tail()
```

```python
print('min:',df['balance'].min(),
     'max:',df['balance'].max())
```

```python
df['balance']=df['balance'].apply(lambda v: (v-df['balance'].min())/(df['balance'].max()-df['balance'].min()))
```

```python
df.head()
```

```python
df.contact.replace({
    'unknown':np.nan,
    'telephone':0,
    'cellular':1
},inplace=True)
```

```python
df.month.replace({
    'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12
},inplace=True)
```

```python
df.month.unique()
```

```python
df.head()
```

```python
df.poutcome.unique()
```

```python
df.poutcome=df.poutcome.map({
    'unknown':0, 
    'failure':1, 
    'other':2, 
    'success':3
})
```

```python
df.pdays=df.pdays.apply(lambda v : (v-df.pdays.min())/(df.pdays.max()-df.pdays.min()))
```

```python
df.head()
```

```python
df.y.replace({
    'no':0,
    'yes':1
},inplace=True)
```

```python
    df.head()
```

```python
df.duration=df.duration.apply(lambda v : ((v-df.duration.min())/(df.duration.max()-df.duration.min())))
```

```python
df.head()
```

```python
#statistical analysis
df.describe()
```
