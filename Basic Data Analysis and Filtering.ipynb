```python
#Import the pandas and numpy modules

import pandas as pd
import numpy as np
```

```python
#Load in the Excel data that represents a yearâ€™s worth of sales for our sample company
df = pd.read_excel("sales_data.xlsx")
```

```python
#Take a quick look at the data types
df.dtypes
```

```python
#Convert 'date' generic object to datetime object to make some future selections easy
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
df.dtypes
```

```python
#Fitering an account number

df[df["account number"]==307599].head()
```

```python
#Filter quantity greater than 22

df[df["quantity"] > 22].head()
```

```python
#Using map function to filter data 
df[df["sku"].map(lambda x: x.startswith('B1'))].head()
```

```python
#Multiple filter using map function
df[df["sku"].map(lambda x: x.startswith('B1')) & (df["quantity"] > 22)].head()
```

```python
#Filter a specific list of values
df[df["account number"].isin([714466,218895])].head()
```

```python
#String function filtering
df[df['sku'].str.contains('B1')].head()
```

```python
#Date filtering
df[df['date'] >='20140905'].head()
```

```python
#Partial Date Filtering
df[df['date'] >='2014-03'].head()
```

```python
#Date Range- Date Chain
df[(df['date'] >='20140701') & (df['date'] <= '20140715')].head()
```

```python
#setting date as index
df2 = df.set_index(['date'])
df2.head()
```

```python
#Data Slicing
df2["20140101":"20140201"].head()
```

```python
#Unique vales
df["name"].unique()
```

```python
#Drop duplictaes and select unique values
df.drop_duplicates(subset=["account number","name"]).ix[:,[0,1]]
```
